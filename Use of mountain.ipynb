```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import mountain as m
```

Producing testdata

```python
x=np.linspace(-50,50,500)
y=np.linspace(-50,50,500)
xg,yg=np.meshgrid(x,y)
z=np.cos(xg*yg)+xg
z1=4*np.exp(-(xg**2+(5+yg)**2)/100)
z2=7*np.exp(-((xg-20)**2+(5+yg)**2)/50)
z3=5*np.exp(-((xg-20)**2+(25+yg)**2)/250)
z4=5*np.exp(-((xg+45)**2+(-25+yg)**2)/450)
z5=4*np.exp(-((xg+25)**2+(-35+yg)**2)/20)

z=z1+z2+z3+z4+z5+np.random.normal(0,0.01,[500,500])
```

Mountfast interprets the input array or the input image as a map with heights at every grid position. The algorithm puts a mountainclimber on every position in the grid and then allows the climbers to take one step per loop. The climbers go to one of the four neighbour positions, if the value at this position is higher than at their current position. If there are two directions, in which it is going up, the climber splits and half of him goes to each higher position. So in contrast to a gradient algorithms all positive slopes (not only the biggest one) are taken into account. To make the algorithm a bit more stable facing noisy data, not only the next neighbour contributes to the decision of a climber to go in this direction, but also the second position from the climber. If the value in a distance of two grid positions from the climber is bigger, the climber can go down for one step. If there is no higher value near to the climber any more, the climber stops. If no climber is moving anymore or the difference caused by the still ongoing movement is very small, the algorithm terminates and returns an array with the same dimensions as the input array. This values of this array represent how many climbers are at every position.

```python
a=m.mountfast(z)
```

Flood does the opposite of Mountfast, it puts on the top of every mountain people, that only go downwards, where all the people coming from one mountain carry the same value. Flood takes two main input arguments, the array which represents the map containing the heights of the mountains and the array representing the number of climbers on every position. The floodthreshold should separate noise from real mountains. Its values represents the area of a mountain. The clustercheckradius serves to merge mountains, since a big (wide) mountain is likely to have noise on him and therefore multiple relative maxima. The clustercheck merges these to one mountain.  

```python
b,c=m.flood(z,a,floodthreshold=100,clustercheckradius=0.45)
```
